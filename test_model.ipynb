```python
import joblib
import numpy as np
import pandas as pd

# Load the saved model and scaler
rf_model = joblib.load("random_forest_fraud_model.pkl")
scaler = joblib.load("scaler.pkl")
```

```python
feature_names = [
    "SpendingPattern1", "TransactionRisk2", "UserBehavior2", "SecurityScore",
    "FrequencyFlag2", "TransactionRisk1", "VendorTypeScore", 
    "SpendingPattern2", "TransactionVolume"
]
```

```python
# Example
custom_input = pd.DataFrame([[
    -0.5,  # SpendingPattern1
    1.3,   # TransactionRisk2
    -0.07, # UserBehavior2
    0.13,  # SecurityScore
    -0.3,  # FrequencyFlag2
    2.5,   # TransactionRisk1
    0.2,   # VendorTypeScore
    -0.6,  # SpendingPattern2
    0.09   # TransactionVolume
]], columns=feature_names)

# Scale and predict
custom_scaled = scaler.transform(custom_input)
pred = rf_model.predict(custom_scaled)
proba = rf_model.predict_proba(custom_scaled)

print("Prediction:", "Fraud" if pred[0] == 1 else "Legit")
print("Fraud Probability:", proba[0][1])
```
