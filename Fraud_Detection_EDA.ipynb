```python
top_features = [
    'SpendingPattern1', 'TransactionRisk2', 'UserBehavior2',
    'SecurityScore', 'FrequencyFlag2', 'TransactionRisk1',
    'VendorTypeScore', 'SpendingPattern2', 'TransactionVolume'
]
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, fmt=".3f", cmap='Blues')
plt.title('Correlation Heatmap of Features')
```

```python
def cap_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    clipped_series = series.clip(lower=lower, upper=upper)
    return clipped_series, lower, upper
```

```python
bounds = []

for col in top_features:
    df[col], lower, upper = cap_outliers_iqr(df[col])
    bounds.append({'Feature': col, 'Lower Bound': lower, 'Upper Bound': upper})
```

```python
bounds
```

```python
for col in top_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='IsFraud', y=col, data=df)
    plt.title(f'{col} vs IsFraud (After IQR Capping)')
    plt.show()
```
